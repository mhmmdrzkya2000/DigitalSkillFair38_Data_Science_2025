# Mengatasi missing value pada kolom age

```python
# Drop duplikat
data = df.drop_duplicates()

# Isi missing value pada kolom 'age' dengan nilai rata-rata
data['age'].fillna(data['age'].mean(), inplace=True)
```

```python
# Cek missing value dan persentasenya
data.isnull().sum()
(data.isnull().sum() / len(df)) * 100
```

```python
for column in data.columns:
    print(f"============= {column} =================")
    display(data[column].value_counts())
    print()
```

Visualisasi Data

```python
df = data

# Visualisasi 1: Distribusi Penumpang Selamat vs Tidak Selamat
fig1 = px.histogram(df, x='survived', title='Distribusi Penumpang: Selamat vs Tidak Selamat', color='survived')

# Visualisasi 2: Distribusi Jenis Kelamin Penumpang
fig2 = px.histogram(df, x='sex', title='Distribusi Jenis Kelamin Penumpang', color='sex')

# Visualisasi 3: Distribusi Umur Penumpang
fig3 = px.histogram(df, x='age', nbins=30, title='Distribusi Umur Penumpang')

# Visualisasi 4: Perbandingan Survival Berdasarkan Jenis Kelamin
fig4 = px.histogram(df, x='sex', color='survived', barmode='group',
                    title='Perbandingan Jenis Kelamin dan Status Keselamatan')

# Dash app layout
app = dash.Dash(__name__)
app.title = "Dashboard Titanic"

app.layout = html.Div([
    html.H1("Dashboard Analisis Titanic", style={'textAlign': 'center'}),

    dcc.Graph(figure=fig1),
    dcc.Graph(figure=fig2),
    dcc.Graph(figure=fig3),
    dcc.Graph(figure=fig4),
])

if __name__ == '__main__':
    app.run(debug=True)
```
